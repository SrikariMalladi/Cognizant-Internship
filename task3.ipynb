```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the datasets
df1 = pd.read_csv("4.1 sample_sales_data.csv")


# Define a function to create visualizations for a given dataset
def visualize_data(df, dataset_name):
    print("Visualizations for", dataset_name)

    # Histogram
    plt.hist(df['unit_price'], bins=10)
    plt.xlabel('Unit Price')
    plt.ylabel('Frequency')
    plt.title('Histogram of Unit Price')
    plt.show()

    # Bar Chart
    plt.figure(figsize=(8, 6))
    sns.countplot(x='category', data=df)
    plt.xlabel('Category')
    plt.ylabel('Count')
    plt.title('Bar Chart of Categories')
    plt.show()

    # Scatter Plot
    plt.scatter(df['unit_price'], df['quantity'])
    plt.xlabel('Unit Price')
    plt.ylabel('Quantity')
    plt.title('Scatter Plot of Unit Price vs Quantity')
    plt.show()

    # Box Plot
    plt.figure(figsize=(8, 6))
    sns.boxplot(x='category', y='quantity', data=df)
    plt.xlabel('Category')
    plt.ylabel('Quantity')
    plt.title('Box Plot of Quantity by Category')
    plt.show()

    # Pie Chart
    plt.figure(figsize=(8, 6))
    df['customer_type'].value_counts().plot(kind='pie', autopct='%1.1f%%')
    plt.title('Pie Chart of Customer Types')
    plt.ylabel('')
    plt.show()

    # Heatmap
    plt.figure(figsize=(10, 8))
    numeric_df = df.select_dtypes(include=['number'])
    sns.heatmap(numeric_df.corr(), annot=True, cmap='viridis', fmt=".2f")
    plt.title('Correlation Heatmap')
    plt.show()

    # Stacked Bar Chart
    plt.figure(figsize=(8, 6))
    df.groupby(['category', 'payment_type']).size().unstack().plot(kind='bar', stacked=True)
    plt.xlabel('Category')
    plt.ylabel('Count')
    plt.title('Stacked Bar Chart of Payment Types by Category')
    plt.show()

    # Violin Plot
    plt.figure(figsize=(8, 6))
    sns.violinplot(x='customer_type', y='unit_price', data=df)
    plt.xlabel('Customer Type')
    plt.ylabel('Unit Price')
    plt.title('Violin Plot of Unit Price by Customer Type')
    plt.show()

# Visualize each dataset
visualize_data(df1, "sample_sales_data.csv")

```
